```python
import pandas as pd
```

```python
df = pd.read_csv('Train.csv')
```

```python
df
```

```python
df.info()
```

```python
# Unique image
df['Image_ID'].nunique()
```

```python
# Group by Image_ID and keep the first occurrence
unique_images_df = df.drop_duplicates(subset=['Image_ID'])

# Check the shape to confirm we have 5529 unique images
print(f"Original dataframe shape: {df.shape}")
print(f"Unique images dataframe shape: {unique_images_df.shape}")
```

```python
# Group by Image_ID and aggregate relevant information
grouped_df = df.groupby('Image_ID').agg({
        'class': list,  # Take first class
        'ImagePath': lambda x: x.iloc[0],  # Take first path
        'confidence': 'mean',  # Average confidence
        'class_id': list,  # Take first class_id
        'ymin': list,  # List of all ymin values
        'xmin': list,  # List of all xmin values
        'ymax': list,  # List of all ymax values
        'xmax': list,  # List of all xmax values
}).reset_index()
```

```python
grouped_df
```

```python
# Create explicit copies before modifying
healthy_images = grouped_df[grouped_df['class'].apply(lambda x: all(c == 'healthy' for c in x))].copy()
other_classes_images = grouped_df[grouped_df['class'].apply(lambda x: any(c != 'healthy' for c in x))].copy()

# Now add the labels
healthy_images['labels'] = "Normal"
other_classes_images['labels'] = "Disease"
```

```python
healthy_images
```

```python
other_classes_images
```

```python

# New dataframe with labels, image paths, and image id
final_df = pd.concat([healthy_images[['Image_ID', 'ImagePath', 'labels']],
                       other_classes_images[['Image_ID', 'ImagePath', 'labels']]], ignore_index=True)
```
